```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = loadtxt('/home/pima-indians-diabetes.csv', delimiter = ',')

#Split into input which is x and output which is the y variable

x = dataset[:,0:8]
y = dataset[:,8]
```

```python
#Define the keras model.
model = Sequential()
#rectified linear unit(relu)
model.add(Dense(12, input_dim= 8, activation = 'relu'))#This is the first layer.
model.add(Dense(8, activation = 'relu'))#Has 8 noodes, this is the second layer.
model.add(Dense(1,activation= 'sigmoid'))#The third layer.
```

```python
#Compile the keras model.
#stochastic gradient function.
model.compile(loss = 'binary_crossentropy', optimizer = 'adam',metrics = ['accuracy'])
```

```python
# Fit the keras model on the dataset.
model.fit(x,y, epochs = 200, batch_size = 10)
```

```python
#Evaluate the keras model.
_, accuracy = model.evaluate(x,y)
print('Accuracy: %.2f' %(accuracy*100))
```
